```python
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2                           #引用各种库
```

```python
img = cv2.imread("lenna.png")
h,w,z = img.shape[:3]             #d读取图片对应矩阵的行、列数量
```

```python
img_gray = np.zeros([h,w],img.dtype)  #创建单通道图片
for i in range(h):
    for j in range(w):
        m = img[i,j]                  #储存【i，j】位置像素点的BGR值
        img_gray[i,j]  = int(m[0]*0.11+m[1]*0.59+m[2]*0.3)   #浮点计算，转换为灰度值
```

```python
# print (m)
# print (img_gray)
# print("image show gray: %s"%img_gray)
cv2.imshow("image show gray",img_gray)
plt.subplot()
img = plt.imread("lenna.png") 
# img = cv2.imread("lenna.png")    #cv2.imread和plt.imread读取的RGB顺序不同
plt.imshow(img)
print("---image lenna----")
```

```python
# 灰度处理
img_gray = rgb2gray(img)
# img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# img_gray = img
plt.subplot()
plt.imshow(img_gray, cmap='gray')
print("---image gray----")
```

```python
# 二值化
img_binary = np.where(img_gray >= 0.5, 1, 0) 
plt.subplot() 
plt.imshow(img_binary, cmap='gray')
```

```python
#底片
img_binary = np.where(img_gray <= 0.5, 1, 0) 
plt.subplot() 
plt.imshow(img_binary, cmap='gray')
```
