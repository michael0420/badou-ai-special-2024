```python
# 面向对象：是一种程序设计的思想，将对象作为程序的基本单元
# 面向过程：就是一系列的函数，将这些功能组合在一起

# 注意：在Python中，所有的数据类型都可以视为对象

class Car:
    # 公有属性
    wheelNum = 2
    color = 'red'
    # 私有属性
    __age = 1
    # 构造器方法
    def __init__(self, newWheelNum, newColor):
        self.wheelNum = newWheelNum
        self.color = newColor
        print('self.wheelNum = {0},self.color = {1}'.format(self.wheelNum,self.color))
    # 方法
    def getCarInfo(self):
        print('车轮子个数：%d，颜色：%s，车龄：%d' % (self.wheelNum, self.color, self.__age))
        
    def run(self):
        print('车在奔跑，目的地：夏威夷')
    
print(Car.wheelNum)
# 创建实例对象
print('------构造方法--------')
BMW = Car(4, 'green')

# 访问公有属性
print('车轮子的数量为：',BMW.wheelNum)

# 访问私有属性
# print('车龄：',BMW.__age)

BMW.price = 400000
print('宝马价格：',BMW.price)

# 调用对象的getCarInfo方法
BMW.getCarInfo()

# 调用对象的run方法
BMW.run()

# 继承、多继承
# 定义一个父类，如下
class Phone:
    # 公有属性
    color = 'white'
    price = 4000
    # 方法
    def getPhoneInfo(self):
        print('手机颜色：{0}，价格：{1}'.format(self.color, self.price))
        
# 定义一个子类
class HuaWei(Car, Phone):
    work = 996
    country = 'china'
    #实例方法
    def getHuaWeiInfo(self):
        print('华为上班时间：',self.work)
    #类方法，用classmethod来进行修饰
    @classmethod
    def getCountry(cls):
        return cls.country
    @classmethod
    def setCountry(cls,country):
        cls.country = country
    #静态方法,不需要额外定义参数，使用类对象引用类属性
    @staticmethod
    def getInfo():
        return '华为所属国家：{0}，上班时间：{1}'.format(HuaWei.country,HuaWei.work)
        
print('-------多继承-------')
obj = HuaWei(4,'yellow')
# 实例属性:不需要在类中显示定义
print('-------实例属性-------------')
obj.country = 'ShenZhen'
print(obj.country)
del obj.country # 删除实例属性
print(obj.country)
print('-------实例方法--------')
obj.getHuaWeiInfo()
obj.getPhoneInfo()
obj.getCarInfo()
print('-------类方法-------')
print(obj.getCountry())
print(HuaWei.getCountry())
print('------静态方法--------')
print(HuaWei.getInfo())
```
