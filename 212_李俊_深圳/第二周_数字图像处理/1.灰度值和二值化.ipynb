```python
from  skimage.color  import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from PIL import  Image
import cv2
```

```python
#灰度化
img = cv2.imread(r"C:\Users\Administrator\Desktop\lenna.png")
#获取图片的宽度和高度
h,w = img.shape[:2]
img_gray = np.zeros([h,w],img.dtype) #创建同样大小的单通道
for i in range(h):
    for j in range(w):        #遍历取出BGR坐标
        m = img[i,j]
        img_gray[i,j] = int(m[0]*0.11+m[1]*0.59+m[2]*0.3) #转换为灰度坐标
print(m)
```

```python
plt.subplot(221)
img = plt.imread(r"C:\Users\Administrator\Desktop\lenna.png")
plt.imshow(img)
print("---image lenna----")
print(img)
```

```python
#灰度化
img_gray = rgb2gray(img)
plt.subplot(222)
plt.imshow(img_gray)
plt.imshow(img_gray,cmap='gray')
print("---image gray----")
print(img_gray)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(222)
plt.imshow(img_gray, cmap='gray')
print("---image gray----")
print(img_gray)
```

```python
#二值化
rows,cols = img_gray.shape
for i in range(rows):
    for j in range(cols):
        if (img_gray[i,j] <= 0.5):
            img_gray[i,j] = 0
        else:
            img_gray[i,j] = 1
print("-----imge_binary------")
print(img_gray)
print(img_gray.shape)
```

```python
img_binary = np.where(img_gray >= 0.5, 1, 0) 
print("-----imge_binary------")
print(img_binary)
print(img_binary.shape)

plt.subplot(223) 
plt.imshow(img_binary, cmap='gray')
plt.show()
```
