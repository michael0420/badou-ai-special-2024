```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
def function(img):
  hight, width, channels = img.shape
  emptyImage = np.zeros([800,800,channels], np.uint8)
  sh = 800/hight
  sw = 800/width
  for i in range(800):
    for j in range(800):
      #int(),转为整型，使用向下取整。
      x = int(i/sh + 0.5)
      y = int(j/sh + 0.5)
      emptyImage[i,j] = img[x,y]
  return emptyImage

# cv2.resize(img, (800,800,c),near/bin)

img=cv2.imread("lenna.png")
zoom=function(img)
print(zoom)
print(zoom.shape)
print("nearest interp")
cv2_imshow(zoom)
print("image")
cv2_imshow(img)
cv2.waitKey(0)
```
